# Windows10 で Graphviz を使うための設定

1.  pydotplus, graphviz の python モジュール
2.  graphviz 本体

が必要です。

---

## 1. pydotplus, graphviz の python モジュールのインストール

スタートメニューから command prompt か　　Anaconda prompt を起動し、次の2行を順に実行して下さい。

<pre>
pip install pydotplus
pip install graphviz
</pre>

---


## 2. graphviz 本体の入手

[Graphviz - Graph Visualization Software](http://www.graphviz.org/Download_windows.php) から入手できます。

管理者になれるのであれば、　msi 版をダウンロードしてインストールしてしまえば終わりです。


---

## 3. 管理者になれない場合の利用法

上記サイトから、zip 版をダウンロードして展開して下さい。どこに置いてもよいですが、デスクトップにおいて使うことを前提に述べます。

 **graphviz-2.xx.zip** を展開すると **release** というフォルダができます（ graphviz-2.xx ができる場合はその中の release ）．その中にbin というフォルダがあるので，  **Graphviz** にリネームし，デスクトップに移動させてください． bin 以外のフォルダは不要です。

### 3-1 再起動で設定がリセットされる環境の場合
大学の全学共用パソコンルームやネットカフェでは再起動ですべての設定や保存物がリセットされるように設定されていることが多いです。そのような場合は、毎回環境変数をセットして使うしかありません。

1回限りであれば、jupyter notebook や python を起動する直前に、次のコマンドを実行する。


---

```
set Path %Path%;%USERPROFILE%\Desktop\Graphviz
```

---

この設定はターミナルを閉じると再度実行する必要がありますし、実行前に動かした notebook には効きません。

面倒ですので次のようなバッチファイルを作りましょう。

---
<pre class="lang:default decode:true">
set Path=%Path%;%USERPROFILE%\Desktop\Graphviz
jupyter notebook</pre>

---

メモ帳を開いてこの２行をコピペし、** runjupyter.bat ** という名前で jupyter notebook を起動したいフォルダに保存してください。　　

ダブルクリックすれば graphviz が使える状態で notebook が起動します。

---

### 3-2  再起動で設定が残せる場合
合同演習室のPCなどでは管理者になれなくても、再起動で設定がリセットされることはありませんので、毎回 Pathを通す必要はありません。その場合は次のようにしてください。

1.  CORTANA(Windows のパーソナルアシスタント、タスクバー左端の検索窓）に「環境変数を編集」と入力する．
２．　ユーザ環境変数が上部に，システム環境変数が下部に表示されるので，上部に Path があるかどうか調べる．
３. あるならそれをダブルクリック、なければ新規に作る．
4. もう一つウィンドウが重なって開く．**新規(N)** をクリックし，** %USERPROFILE%\Desktop\Graphviz **というエントリーを追加．OK,OK　で適用して閉じる。
5. 再起動
